```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

tweets_list = []

query = "(#bist100 OR #bist30 OR #xu100 OR #piyasa OR #endeks) until:2023-06-06 since:2023-01-01"

for i, tweet in enumerate(sntwitter.TwitterSearchScraper(query).get_items()):
    try:
        if not str(tweet.id).startswith("promoted-tweet") and tweet.viewCount is not None and tweet.viewCount >= 50000:
            tweets_list.append([tweet.date, tweet.rawContent, tweet.viewCount])
    except AttributeError:
        continue

    if i % 1000 == 0:  # Print the count every 1000 tweets
        print("Downloaded tweets:", i)

print("Total tweets collected:", len(tweets_list))

```

```python
# Creating a dataframe from the tweets list above
tweets_df2 = pd.DataFrame(tweets_list, columns=['Datetime', 'Text', 'View'])
tweets_df2.to_csv("parsed03.csv", index = False)
```
